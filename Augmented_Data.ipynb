```python
import Augmentor
import os

folder_names = ['Autoimun_Dermatomyositis', 'Autoimun_Vitiligo', 'Autoimun_Psoriasis', 'Autoimun_Lichen_Planus', 'Nonautoimun_Tinea', 'Nonautoimun_Eczema', 'Nonautoimun_Herpes', 'Nonautoimun_Keratosis_Seborrheic']
base_dir = '/content/drive/My Drive/DATASET/source'
augmented_dir_ = '/content/drive/My Drive/DATASET/output/augmented'  # Renamed to avoid conflict

for folder_name in folder_names:
  source_dir = os.path.join(base_dir, folder_name)
  output_dir = os.path.join(output_dir, folder_name) # Using the base output directory

  # Create the output directory if it doesn't exist
  os.makedirs(output_dir, exist_ok=True)

  # The output_dir is no longer passed to the Pipeline constructor
  p = Augmentor.Pipeline(source_directory=source_dir, output_directory=output_dir, save_format="jpg")
  p.rotate(probability = 0.7, max_left_rotation = 10, max_right_rotation = 10)
  p.flip_left_right(probability = 0.5)
  p.flip_top_bottom(probability = 0.5)
  p.zoom_random(probability = 0.5, percentage_area = 0.8)
  p.crop_random(probability = 0.5, percentage_area = 0.8)
  p.resize(probability = 1.0, width = 224, height = 224)
  p.random_brightness(probability = 0.5, min_factor = 0.8, max_factor = 1.2)
  p.random_contrast(probability = 0.5, min_factor = 0.8, max_factor = 1.2)
  p.sample(1000)
```

```python
import os
import pandas as pd
import shutil

# Path direktori gambar
base_dir = '/content/drive/My Drive/DATASET/source'
augmented_dir = '/content/drive/My Drive/DATASET/output/augmented'

# Membuat file CSV untuk mencatat nama gambar dan label
csv_file = os.path.join(augmented_dir, 'labels_augmented.csv')

# Inisialisasi list untuk menyimpan data CSV
labels_data = []

# Nama kolom untuk DataFrame
columns = ['images', 'Autoimun_Dermatomyositis', 'Autoimun_Vitiligo', 'Autoimun_Psoriasis', 'Autoimun_Lichen_Planus',
           'Nonautoimun_Tinea', 'Nonautoimun_Keratosis_Seborrheic', 'Nonautoimun_Eczema', 'Nonautoimun_Herpes']

# Membaca semua subdirektori di dalam direktori gambar (setiap subdirektori adalah label)
for class_name in os.listdir(base_dir):
    class_dir = os.path.join(base_dir, class_name)

    # Pastikan hanya membaca folder, bukan file
    if os.path.isdir(class_dir):

        # Membuat folder untuk label jika belum ada di output_dir
        class_output_dir = os.path.join(augmented_dir, class_name)
        os.makedirs(class_output_dir, exist_ok=True)

        # Iterasi gambar-gambar dalam folder tersebut
        for image_name in os.listdir(class_dir):
            image_path = os.path.join(class_dir, image_name)

            # Pastikan hanya file gambar yang diproses (misal .jpg atau .jpeg)
            if os.path.isfile(image_path) and image_name.lower().endswith(('.jpg', '.jpeg')):

                # Buat dictionary untuk setiap baris data
                row_data = {col: 0 for col in columns}  # Inisialisasi semua kolom dengan 0
                row_data['images'] = image_name
                row_data[class_name] = 1  # Set nilai 1 untuk kolom kelas yang sesuai

                # Tambahkan data baris ke list
                labels_data.append(row_data)

                # Pindahkan gambar ke direktori output sesuai label
                shutil.copy(image_path, os.path.join(class_output_dir, image_name))

# Membuat DataFrame dari data yang dikumpulkan dan simpan ke CSV
df = pd.DataFrame(labels_data, columns=columns)
df.to_csv(csv_file, index=False)

print(f'File CSV telah disimpan di {csv_file}')
```
